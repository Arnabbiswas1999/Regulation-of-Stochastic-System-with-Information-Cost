```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
A_0 = np.random.rand(4,1)
A_0 = A_0 / A_0.sum(axis = 0)[:, None]
```

```python
H_0 = np.random.rand(16,16)
H_0 = H_0 / H_0.sum(axis = 1)[:, None]
H_1 = np.random.rand(16,16)
H_1 = H_1 / H_1.sum(axis = 1)[:, None]
H_2 = np.random.rand(16,16)
H_2 = H_2 / H_2.sum(axis = 1)[:, None]
H_3 = np.random.rand(16,16)
H_3 = H_3 / H_3.sum(axis = 1)[:, None]
```

```python
mat=A_0[0][0]*H_0+A_0[1][0]*H_1+A_0[2][0]*H_2+A_0[3][0]*H_3
A=mat
A_n = A
```

```python
Y=np.random.rand(16,1)
#Y = np.array([[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625],[0.0625]])
x = np.linspace(1, 16, num=16)
Y = Y.T
for i in range(10):
    
    Y = np.matmul( Y, A_n)
    y_temp = np.squeeze(Y)
    plt.plot(x, y_temp, ( 'C' + str(i + 1)), label = ( 'C' + str(i + 1)))
    plt.legend()
    if i == 9:
      plt.show()
    
```
