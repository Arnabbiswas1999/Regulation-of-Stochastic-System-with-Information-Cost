```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
mat = np.random.rand(4, 16)
mat = mat / mat.sum(axis = 1)[:, None]
mat=np.array(mat)
print(mat)
```

```python
result=mat.reshape(1,64)
print(result)
```

```python
result.shape
```

```python
column_mat=result.transpose()
print(column_mat)
```

```python
column_mat.shape
```

```python
arr= np.random.rand(16,64)
arr=arr/10
print(arr)
```

```python
final=np.matmul(arr, column_mat)
print(final)
```

```python
from itertools import chain
final_list = list(chain.from_iterable(final))
print(final_list)
```

```python
size = len(final_list)
plt.plot(final_list)
```
