```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
H_mat = np.random.rand(2,2)
H_mat = H_mat / H_mat.sum(axis = 1)[:, None]
H_mat=np.array(H_mat)
print(H_mat)  
```

```python
Y_0=np.random.rand(16,1)
A_1=np.random.rand(1,4)
```

```python
mat=np.matmul(Y_0,A_1)
```

```python
result=mat.reshape(1,64)
```

```python
column_mat=result.transpose()
```

```python
Y_1=np.matmul(H_mat,column_mat)
print(Y_1)
```

```python
A_2=np.random.rand(1,4)
mat=np.matmul(Y_1,A_2)
result=mat.reshape(1,64)
column_mat=result.transpose()
Y_2=np.matmul(H_mat,column_mat)
print(Y_2)
```
