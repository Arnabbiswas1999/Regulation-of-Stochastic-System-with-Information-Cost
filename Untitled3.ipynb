This code is implemented to find the steady state matrix of P(x). We assume a transition probability matrix where the probability of each state is same. After finding these two matrices, we find the entropy of system which we generate in the previous implementation.

```python
import numpy as np
transition_matrix=np.random.rand(16,16)
```

```python
steps=341
transition_matrix_n=transition_matrix
i=0
while i<steps:
  transition_matrix_n=np.matmul(transition_matrix_n,transition_matrix)
  i+=1
print(transition_matrix_n[0])
```

```python
import scipy.linalg
values, left = scipy.linalg.eig(transition_matrix, right = False, left = True)

#print("left eigen vectors = \n", left, "\n")
print("eigen values = \n", values)

```

```python
pi = left[:,8]
pi_normalized = [(x/np.sum(pi)).real for x in pi]
pi_normalized
```

```python
s=0
for i in range(16) :
  s = s + pi_normalized[i]

#sum=sum*pi_normalized
print(s)
```

```python
sum = 0
for i in range(16) :
  sum = sum + (transition_matrix[8][i] * np.log(transition_matrix[8][i])) * pi_normalized[i]

#sum=sum*pi_normalized
print(sum)
  
```
